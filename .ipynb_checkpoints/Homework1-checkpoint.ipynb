```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import confusion_matrix
```

## (a) Vertebral Column Data Set: https://archive.ics.uci.edu/ml/datasets/Vertebral+Column

## (b) Pre-Processing and Exploratory Data Analysis

```python
columns = ["Pelvic Incidence", 'Pelvic Tilt','Lumbar Lordosis Angle', 'Sacral Slope', 'Pelvic Radius' ,'Degree Spondylolisthesis', 'Class']
df = pd.read_csv("column_2C.dat",sep=" ",header=None, names=columns)
df.head()
```

### i. Make scatterplots of the independent variables in the dataset.

```python
ind_variables = ["Pelvic Incidence", 'Pelvic Tilt','Lumbar Lordosis Angle', 'Sacral Slope', 'Pelvic Radius' ,'Degree Spondylolisthesis']
plt.figure(figsize = (6,6))
sns.pairplot(df, y_vars = ind_variables, x_vars = ind_variables, hue = "Class", plot_kws=dict(s=20))
plt.show()
```

### ii. Make boxplots for each of the independent variables.

```python
ind_variables = ["Pelvic Incidence", 'Pelvic Tilt','Lumbar Lordosis Angle', 'Sacral Slope', 'Pelvic Radius' ,'Degree Spondylolisthesis']
plt.figure(figsize=(12,20))
subplt = 1
for i in range(0,len(ind_variables)):
    plt.subplot(3,2, subplt)
    sns.boxplot(x = "Class", y = ind_variables[i], data = df)
    subplt += 1
```

### iii. Select the training and test data set

```python
df['Class Label'] = df.apply(lambda x: 0 if x['Class'] == 'NO' else 1, axis=1)
df.columns = ["Pelvic Incidence", 'Pelvic Tilt','Lumbar Lordosis Angle', 'Sacral Slope', 'Pelvic Radius' ,'Degree Spondylolisthesis', 'Classification', 'Class']
              
train_normal = df[df['Class'] == 0].iloc[0:70,:]
test_normal = df[df['Class'] == 0].iloc[70:,:]
train_abnormal = df[df['Class'] == 1].iloc[0:140,:]
test_abnormal = df[df['Class'] == 1].iloc[140:,:]

train_df = pd.concat([train_normal,train_abnormal], axis=0).reset_index(drop=True)
test_df = pd.concat([test_normal,test_abnormal], axis=0).reset_index(drop=True)
```

## (c) Classification using KNN on Vertebral Column Data Set

### ii. Test all the data in the test database with k nearest neighbors.

```python
X_train = train_df.iloc[:,0:6]
y_train = train_df.Class
X_test = test_df.iloc[:,0:6]
y_test = test_df.Class

train_errors = []
test_errors = []
k_values = list(range(1,209,3))

for k in k_values:
    knn = KNeighborsClassifier(k, metric = "euclidean")
    knn.fit(X_train, y_train)
    train_error = 1 - knn.score(X_train,y_train)
    test_error = 1 - knn.score(X_test,y_test)
    train_errors.append(train_error)
    test_errors.append(test_error)

plt.plot(k_values,train_errors, color = 'orange',label = 'Train')
plt.plot(k_values,test_errors, color = 'blue',label = 'Test')
plt.xlabel('K Nearest Neighbors')
plt.ylabel('Error Rate')
plt.title('Error Rate Under Varying K Nearest Neighbors')
plt.legend(loc='upper right')
plt.gca().invert_xaxis()
```

```python
index = test_errors.index(min(test_errors))
print('The most suitable k among the values is %d with a test error of %s.'%(k_values[index], round(test_errors[index],3)))

lowest_df = pd.DataFrame(columns=['Distance', 'Training Error'])
lowest_df.loc[0] = ['Euclidean 1', round(min(train_errors),3)]

knn = KNeighborsClassifier(4, metric="euclidean")
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print('\nThe confusion matrix for test data is: ')
                    
data = [[tp, fn], [fp, tn]]
confusion_df = pd.DataFrame(data, columns=["Predicted Abnormal", "Predicted Normal"])
confusion_df = confusion_df.rename(index={0: 'Actual Abnormal', 1: 'Actual Normal'})
print(confusion_df)

true_pos = tp / (tp + fn)
true_neg = tn / (tn + fp)
print('\nThe true positive rate is', round(true_pos, 4))
print('The true negative rate is', round(true_neg, 4))


precision = tp / (tp + fp)
print('The precision is', round(precision, 4))

recall = tp / (tp + fn)
f_score = 2* ((precision * recall) / (precision + recall))
print('The F-score is', round(f_score, 4))
```

### iii. Vary size of training set

```python
training_size = list(range(10,211,10))
training_rate = []
test_rate = []

for i in training_size:
    train_normal = train_df[train_df['Class'] == 0].iloc[0:(i//3)]
    train_abnormal = train_df[train_df['Class'] == 1].iloc[0:(i-i//3)]

    train_df_2 = pd.concat([train_normal,train_abnormal], axis=0).reset_index(drop=True)
 
    X_train = train_df_2.iloc[:,0:6]
    y_train = train_df_2.Class
    X_test = test_df.iloc[:,0:6]
    y_test = test_df.Class

    knn = KNeighborsClassifier()
    train_errors=[]
    test_errors=[]
    k_values = list(range(1,i+1,5))
    for k in k_values:
        knn.n_neighbors=k 
        knn.fit(X_train,y_train)
        train_error = knn.score(X_train, y_train)
        test_error = knn.score(X_test, y_test)
        train_errors.append(train_error)
        test_errors.append(test_error)
    
    index = test_errors.index(max(test_errors))    
    training_rate.append(1-train_errors[index])
    test_rate.append(1-max(test_errors))

index = test_rate.index(min(test_rate))
lowest_df.loc[1] = ['Euclidean 2', round(min(training_rate),3)]
    
plt.plot(training_size,test_rate, color='blue',label='Training Accuracy')
plt.xlabel('Training Set Size')
plt.ylabel('Test Error Rate')
plt.title('Error Rate Under Varying Training Set Size')   
```

## (d) Replace the Euclidean metric with the following metrics and test them.

```python
# Table to summarize test errors
distances_df = pd.DataFrame(columns = ['Distance', 'K Value', 'Test Error'])
```

### iA. Minkowski Distance which becomes Manhattan Distance with p = 1

```python
k_values = list(range(1,197,5))
train_errors = []
test_errors = []

for k in k_values:
    knn = KNeighborsClassifier(k, metric ='manhattan')
    knn.fit(X_train, y_train)
    train_error = 1 - knn.score(X_train, y_train)
    test_error = 1 - knn.score(X_test, y_test)
    train_errors.append(train_error)
    test_errors.append(test_error)

index = test_errors.index(min(test_errors)) 
k = k_values[index]
distances_df.loc[0] = ['Manhattan', 'k = ' + str(k), min(test_errors)]
lowest_df.loc[3] = ['Manhattan', round(min(train_errors),3)]
```

### iB. Minkowski Distance with log10(p)  {0.1,0.2,0.3,...,1}

```python
values = np.linspace(0.1,1,10)
k = 6
train_errors = []
test_errors = []
for i in values:
    power = 10**i
    knn = KNeighborsClassifier(k, algorithm = 'brute', p = power, metric ='minkowski')
    knn.fit(X_train, y_train)
    train_error = 1 - knn.score(X_train, y_train)
    test_error = 1 - knn.score(X_test, y_test)
    train_errors.append(train_error)
    test_errors.append(test_error)

index = test_errors.index(min(test_errors))
distances_df.loc[1] = ['Minkowski','k = ' + str(k), round(min(test_errors),3)]
lowest_df.loc[4] = ['Minkowski', round(min(train_errors),3)]
```

### iC. Minkowski Distance which becomes Chebyshev Distance with p  

```python
k_values = list(range(1,197,5))
train_errors = []
test_errors = []

for k in k_values:
    knn = KNeighborsClassifier(k, metric ='chebyshev')
    knn.fit(X_train,y_train)
    train_error = 1 - knn.score(X_train,y_train)
    test_error = 1 - knn.score(X_test,y_test)
    train_errors.append(train_error)
    test_errors.append(test_error)

index = test_errors.index(min(test_errors))
distances_df.loc[2] = ['Chebyshev', 'k = ' + str(k_values[index]), round(min(test_errors),3)]
lowest_df.loc[5] = ['Chebyshev', round(min(train_errors),3)]
```

### ii. Mahalanobis Distance

```python
k_values = list(range(1,197,5))
train_errors = []
test_errors = []

for k in k_values:
    knn = KNeighborsClassifier(k, algorithm='brute', metric ='mahalanobis', metric_params={'V': np.cov(X_train)})
    knn.fit(X_train, y_train)
    train_error = 1 - knn.score(X_train, y_train)
    test_error = 1 - knn.score(X_test, y_test)
    train_errors.append(train_error)
    test_errors.append(test_error)

index = test_errors.index(min(test_errors))
distances_df.loc[3] = ['Mahalanobis', 'k = ' + str(k_values[index]), round(min(test_errors),3)]
lowest_df.loc[6] = ['Mahalanobis', round(min(train_errors),3)]
```

## (e) Use weighted voting with Euclidean, Manhattan, and Chebyshev distances

```python
weighted_df = pd.DataFrame(columns=['Distance', 'K Value', 'Test Error'])

k_values = list(range(1,197,5))
distances = ['euclidean', 'manhattan', 'chebyshev']
i = 0
for distance in distances:
    train_errors = []
    test_errors = []
    k_values = list(range(1,197,5))
    
    for k in k_values:
        k_nn = KNeighborsClassifier(k, weights='distance', metric=distance)
        k_nn.fit(X_train, y_train)
        train_error = 1 - k_nn.score(X_train, y_train)
        test_error = 1 - k_nn.score(X_test, y_test)
        train_errors.append(train_error)
        test_errors.append(test_error)
    index = test_errors.index(min(test_errors))
    weighted_df.loc[i] = ['Weighted '+ distance.capitalize(), 'k = ' + str(k_values[index]), round(min(test_errors),3)]
    lowest_df.loc[i+8] = ['Weighted '+ distance.capitalize(), round(train_errors[index],3)]
    i += 1

weighted_df
```
