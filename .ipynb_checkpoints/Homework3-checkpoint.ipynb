### ii. Extract the time-domain features minimum, maximum, mean, median, standard deviation, first quartile, and third quartile for all of the 6 time series in each instance. 

```python
column = []
features = ['min', 'max', 'mean', 'median', 'std', '1st quart', '3rd quart']

for num in range(1,7):
    for i in features:
        column.append(i + str(num))

feature_df = pd.DataFrame(columns=column)
feature_df.head()
```

```python
num_file = {'bending1':7, 'bending2':6, 'cycling':15, 'lying':15, 'sitting':15, 'standing':15, 'walking':15}
data_dict = {}

i = 0
for key,value in list(num_file.items()):
    for length in range(1, value + 1):
        file = './AReM/%s/dataset%d.csv'%(key, length)

        if file == './AReM/bending2/dataset4.csv':
            df = pd.read_csv(file, sep=' ', skiprows = 5, names = ['time', 'avg_rss12', 'var_rss12', 'avg_rss13',\
                                                                   'var_rss13', 'avg_rss23', 'var_rss23'], \
                                                                     usecols=range(7))
        else:
            df = pd.read_csv(file, sep=',', skiprows = 4, usecols=range(7))
            
        if 'bending' in key:
            df['activity'] = 1
        else:
            df['activity'] = 0

        if key not in data_dict:
            data_dict[key] = [df]
        else:
            data_dict[key].append(df)
        
        instance = []

        for column in range(1,7):
            minimum = df.iloc[:,column].to_frame().min()[0]
            maximum = df.iloc[:,column].to_frame().max()[0]
            mean = df.iloc[:,column].to_frame().mean()[0]
            median = df.iloc[:,column].to_frame().median()[0]
            std = df.iloc[:,column].to_frame().std()[0]
            quartile1 = df.iloc[:,column].to_frame().quantile(.25, axis = 0)[0]
            
            quartile3 = df.iloc[:,column].to_frame().quantile(.75, axis = 0)[0]

            instance.extend((minimum, maximum, mean, median, std, quartile1, quartile3))
        feature_df.loc[i] = instance   
        i += 1

print(feature_df)
```

### iii. Break each time series in your training set into l ∈ {1, 2, . . . , 20} time series of approximately equal length and use logistic regression to solve the binary classification problem 

I chose to use backward selection using sklearn feature selection instead of calculating p-values.

The correct way to do Cross Validation is to make feature selection a part of training model. The selected features are only using the training data, providing a more accurate test error since there is no relationship between those predictors and outcome.

The wrong way is to do select predictors using all the data, then use cross validation for training and testing. This would result in a low test error because the predictors would have been selected for their association with the outcome.

http://www.uvm.edu/~rsingle/stat295/F17/extra/Cross-Validation_%20The%20Right%20and%20Wrong%20Way.pdf

```python
selection_df = pd.DataFrame(columns=['Number of Time Series(l)', 'Number of Features (p)', \
                                     'Cross Validation Score'])

length = list(range(1,21))
index = 0
for l in length:
    train_dict = {}
    loc = 0

    for df in train:
        instance = []
        row = int(df.shape[0] / l)
        for i in range(0,l):
            temp_df = df.iloc[i*row: (i+1)*row] 
            
            for column in range(1,7):
                minimum = temp_df.iloc[:,column].to_frame().min()[0]
                maximum = temp_df.iloc[:,column].to_frame().max()[0]
                mean = temp_df.iloc[:,column].to_frame().mean()[0]
                median = temp_df.iloc[:,column].to_frame().median()[0]
                std = temp_df.iloc[:,column].to_frame().std()[0]
                quartile1 = temp_df.iloc[:,column].to_frame().quantile(.25, axis = 0)[0]
                quartile3 = temp_df.iloc[:,column].to_frame().quantile(.75, axis = 0)[0]
                instance.extend((minimum, maximum, mean, median, std, quartile1, quartile3))

        train_dict[loc] = instance
        loc +=1
    train_df = pd.DataFrame.from_dict(train_dict, orient='index')
    train_df['activity'] = [1 if i < 10 else 0 for i in range(1, 70)] 
    
    X_train = train_df.iloc[:,:-1]
    y_train = train_df[['activity']]

    lr = LogisticRegression()
    rfecv = RFECV(estimator = lr, cv = StratifiedKFold(5))
    rfecv.fit(X_train, y_train.values.ravel())
    selection_df.loc[index] = [l, rfecv.n_features_, max(rfecv.grid_scores_)] 
    
    index += 1
    
selection_df
```

The best value of the pair (l, p) is at l = 3 with 4 features since it has the highest cross validation score of 1.0.

### iv. Report the confusion matrix and show the ROC and AUC for your classifier on train data. 

```python
l = 3
train_dict = {}
loc = 1

for df in train:
    instance = []
    row = int(df.shape[0] / l)

    for i in range(0,l):
        temp_df = df.iloc[i*row: (i+1)*row]              

        for column in range(1,7):
            minimum = temp_df.iloc[:,column].to_frame().min()[0]
            maximum = temp_df.iloc[:,column].to_frame().max()[0]
            mean = temp_df.iloc[:,column].to_frame().mean()[0]
            median = temp_df.iloc[:,column].to_frame().median()[0]
            std = temp_df.iloc[:,column].to_frame().std()[0]
            quartile1 = temp_df.iloc[:,column].to_frame().quantile(.25, axis = 0)[0]
            quartile3 = temp_df.iloc[:,column].to_frame().quantile(.75, axis = 0)[0]
            instance.extend((minimum, maximum, mean, median, std, quartile1, quartile3))
    train_dict[loc] = instance
    loc +=1
train_df = pd.DataFrame.from_dict(train_dict, orient='index')
train_df['activity'] = [1 if i < 10 else 0 for i in range(1, 70)] 

X_train = train_df.iloc[:,:-1]
y_train = train_df[['activity']]

cols = []
for i in range(0,X_train.shape[1]):
    cols.append('col'+str(i))
X_train.columns = cols

lr = LogisticRegression()
rfecv = RFECV(estimator = lr, cv=StratifiedKFold(5), scoring = 'accuracy')
rfecv.fit(X_train, y_train.values.ravel())
        
attr_index = []

for i, value in enumerate(rfecv.support_):
    if value == True:
        attr_index.append(i)
        
new_X_train = X_train.iloc[:,attr_index]
model = lr.fit(new_X_train, y_train.values.ravel())
y_pred = model.predict(new_X_train)

tn, fp, fn, tp = confusion_matrix(y_train, y_pred).ravel()
print('\nThe confusion matrix is: ')
data = [[tp, fn], [fp, tn]]
confusion_df = pd.DataFrame(data, columns=["Predicted Bending", "Predicted No Bending"])
confusion_df = confusion_df.rename(index={0: 'Actual Bending', 1: 'Actual No Bending'})
print(confusion_df)

fpr, tpr, thresholds = roc_curve(y_train, y_pred)
roc_auc = auc(fpr,tpr)
print("\nThe ROC curve for l = 1 is as follows, and the AUC = %.2f"% auc(fpr,tpr))
plt.figure()
plt.plot(fpr, tpr, color='darkorange', label='AUC = %0.2f' % roc_auc)
plt.legend(loc="lower right")
plt.plot([0, 1], [0, 1], color='navy',  linestyle='--')
plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.show()
```

```python
# Train model for best l
l = 1
train_dict = {}
loc = 0

for df in train:
    instance = []
        
    # Normalize the data
    min_max_scaler = preprocessing.MinMaxScaler()
    values_scaled = min_max_scaler.fit_transform(df.values)
    df = pd.DataFrame(values_scaled)
    df.columns = ['time','avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23','activity']
        
    row = int(df.shape[0] / l)
    for i in range(0,l):
        temp_df = df.iloc[i*row: (i+1)*row] 
            
        for column in range(1,7):
            minimum = temp_df.iloc[:,column].to_frame().min()[0]
            maximum = temp_df.iloc[:,column].to_frame().max()[0]
            mean = temp_df.iloc[:,column].to_frame().mean()[0]
            median = temp_df.iloc[:,column].to_frame().median()[0]
            std = temp_df.iloc[:,column].to_frame().std()[0]
            quartile1 = temp_df.iloc[:,column].to_frame().quantile(.25, axis = 0)[0]
            quartile3 = temp_df.iloc[:,column].to_frame().quantile(.75, axis = 0)[0]
            instance.extend((minimum, maximum, mean, median, std, quartile1, quartile3))
    train_dict[loc] = instance
    loc +=1
train_df = pd.DataFrame.from_dict(train_dict, orient='index')
train_df['activity'] = [1 if i < 10 else 0 for i in range(1, 70)] 

X_train = train_df.iloc[:,:-1]
y_train = train_df[['activity']]

# Run Logisitic Regression with Stratified KFold CV using Lasso (L1 penalty)
lr = LogisticRegressionCV(cv = StratifiedKFold(5), penalty = "l1", solver = "liblinear")
lr.fit(X_train, y_train.values.ravel())
print('l=%s  %f' %(l,lr.scores_[1].max()))
c = lr.C_
```
